```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x1 = np.random.exponential(1)

lbda = np.linspace(0,2,100)

L = lbda*np.exp(-lbda*x1)

m=max(L)

lbda[np.where(L == max(L))]

fig, ax = plt.subplots()
ax.plot(lbda, L, )
ax.scatter(lbda[np.where(L == max(L))], m, color = 'red')
plt.show()
```

```python
#x_1 = 1 to 5
np.random.seed(0)
x1 = np.random.exponential(1,5)
leading = lbda**5
exp =0
for i in range(0,4):
    exp += -lbda*x1[i]
L = leading*np.exp(exp)

m=max(L)
fig, ax = plt.subplots()
ax.plot(lbda, L)
ax.scatter(lbda[np.where(L == max(L))], m, color='red')
plt.show()
```

It gets closer to the true lambda 1 as we increase the sample size. 

```python
# x_1 1 to 10
np.random.seed(0)
x1 = np.random.exponential(1,10)
leading = lbda**len(x1)
exp=0
for i in range(0,len(x1)):
    exp += -lbda*x1[i]
L = leading*np.exp(exp)

m=max(L)
fig, ax = plt.subplots()
ax.plot(lbda, L)
ax.scatter(lbda[np.where(L == max(L))], m, color='red')
plt.show()
```

When we computed 10 samples of x, we are getting closer to the true lambda. 

```python
# x_1- 1 to 100
np.random.seed(1)
x1 = np.random.exponential(1,100)
leading = lbda**len(x1)
exp=0
for i in range(0,len(x1)):
    exp += -lbda*x1[i]
L = leading*np.exp(exp)

m=max(L)
fig, ax = plt.subplots()
ax.plot(lbda, L)
ax.scatter(lbda[np.where(L == max(L))], m, color = 'red')
plt.show()
```

The likelihood function get closes and closer to the true lambda 1. If we were to increase the number of samples we would get so close the true lambda that it would look orthagonal to the x axis.
