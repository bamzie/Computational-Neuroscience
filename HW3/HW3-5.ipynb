```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt


#Theta from 0.1 to 2
thta = list(np.linspace(0.1,2,100))
#The values of theta that we are looking for 
key = [0.1, 0.5, 1, 1.5, 2]

#Extend that to theta
thta.extend(key)



rnd.rand()

alf = 1    # membrane time constant (ms)\
I = 0.95       # capacitance scaled input current (mA/F)
th = 1 # spiking threshold (mV)
s = 0.1      # noise standard deviation

T = 1000      # total time to run
dt = 0.01   # time step
nt = int(np.round(T/dt)+1)     # number of entries in vector array
tvec = np.linspace(0,T,nt)     # time vector (ms)

# Create a dictionary of thta keys to store mean spike times for [0.1,0.5,..]
mean_spikes = dict.fromkeys(sorted(thta)) 

# Create a dictionary of thta keys to store var spike times for [0.1,0.5,..]
var_spikes = dict.fromkeys(sorted(thta))

#Looping through thta
for theta in thta:
    st = []
    u = np.zeros(nt)
    #The neuron spike equation
    for j in np.arange(nt-1):
        u[j+1] = u[j]+dt*(1)+np.sqrt(2*dt)*s*rnd.randn();
        #When we are above a thta(i.e 0.1, 0.5,...) record the number of spikes
        if u[j+1]>theta:
            u[j+1]=0;                   # reset the voltage to resting potential
            st = np.append(st,tvec[j+1])    # add on another spike time        
    #Difference in spike times
    isi = np.diff(st)
    
    #mean of isi
    mst = np.mean(isi)
    
    #var of isi
    var = np.var(isi)
    
    #store that in our dictionary 
    mean_spikes[theta] = mst
    var_spikes[theta] = var

#Find the crit pts that spiked at [0.1, 0.5, ....]    
crit_mean_pts = []
crit_var_pts = []
for i in mean_spikes.keys(): 
    if i in key: 
        crit_mean_pts.append(mean_spikes[i])
        
for i in var_spikes.keys(): 
    if i in key: 
        crit_var_pts.append(var_spikes[i])

#Plot    
fig, ax = plt.subplots()
ax.plot(list(mean_spikes.keys()), list(mean_spikes.values()))
ax.set_title("Mean Plot")
ax.scatter(key,crit_mean_pts, color='red')

fig, ax2 = plt.subplots()
ax2.plot(list(var_spikes.keys()), list(var_spikes.values()))
ax2.set_title("Variance Plot")
ax2.scatter(key, crit_var_pts, color = 'red')
plt.show()
```

My theory in part 3b mathces my Variance Plot. That is understandable because as our threshold increases the noise as more of an effect to the spike. So we see the spike times between spikes varies more as our threshold increases.
