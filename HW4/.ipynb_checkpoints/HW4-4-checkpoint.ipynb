## Part A

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

# time vector
T = 5
tvec = np.linspace(0, T, 400)
u1 = np.linspace(-2,2, 400)
u2 = np.linspace(-2,2, 400)


def computeDerivatives(u, t0): 
    #u1 and u2 variables
    u1= u[0];
    u2 = u[1];
    #du1/dt
    du1 = 1 - u1 - u2 + 1 - np.heaviside(t0-1, 1); 
    du2 = 1 - u1 - u2
    return (du1, du2);   

#State of u1 and u2
U = [0.5,0.5]

#Line attractor
lineAttractor = [1 - u1[t] for t,_ in enumerate(tvec)]

#Solve ODE Model 
UI = odeint(computeDerivatives, U, tvec)

fig,axes = plt.subplots()
axes.plot(UI[:,0],UI[:,1],label = "Input" ) 
axes.plot(u2, lineAttractor,'g-', label="Line Attractor")
plt.legend()
plt.show()
```

What is the long term value of $(u_1(t),u_2(t))$?
Does it match your prediction from 2b?

The long term values of $u_1$ and $u_2$ converge to the line attractor, to the point $\frac{1}{2}$. That matches our prediction of $2b$, where $u_1 = u_2 = \frac{1}{2}$

What happens in the long time limit? Why do you get such qualitatively different behavior
than the case studied in (a).

Similarly to b in the long time limit we return to the line attractor, however, $u_1$ and $u_2$ do not converge to $\frac{1}{2}$.
```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

# time vector
T = 5
tvec = np.linspace(0, T, 400)
u1 = np.linspace(-2,2, 400)
u2 = np.linspace(-2,2, 400)
gamma = 1

def computeDerivatives(u, t0): 
    #u1 and u2 variables
    u1= u[0];
    u2 = u[1];
    #du1/dt
    du1 = -u1 + 1/(1 + np.exp(-gamma*(1-2*u2))); 
    du2 = -u2 + 1/(1 + np.exp(-gamma*(1-2*u1))); 
    return (du1, du2);   

#State of u1 and u2
U = [1.0,0.0]


#Solve ODE Model 
UI = odeint(computeDerivatives, U, tvec)

#u1 nullcline and u2 nullclines
u1Null = [1/(1 + np.exp(-gamma + 2*gamma*u)) for u in u2]
u2Null = [1/(1 + np.exp(-gamma + 2*gamma*u)) for u in u1]


fig,axes = plt.subplots()
axes.plot(u1, u1Null, label = "u1", linewidth=2 ) 
axes.plot(u2, u2Null,'g-', label="u2")
plt.legend()
plt.show()
```
