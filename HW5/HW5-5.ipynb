## Part A ##

```python
import numpy as np
import matplotlib.pyplot as plt

T = 10      # total time to run
ts = 1  # flag for patter switch 
dt = 0.001   # time step
nt = int(np.round(T/dt)+1)     # number of entries in vector array
tvec = np.linspace(0,T,nt)     # time vector (ms)

w1 = np.zeros(nt)   # vector of w_1
w2 = np.zeros(nt)  # vector of w_2 
w1[0] = 1; 
w2[0] = 1

v_theta = 2

flag = 1 
pattern = 0

for j in np.arange(1,nt-1):
    if flag: 
        eta = 1; 
        u1 = 3; 
        u2 = 0; 
        
    else: 
        eta = 2; 
        u1 = 0; 
        u2 = 1; 
    
    v = w1[j-1]*u1 + w2[j-1]*u2
    
    w1[j] = w1[j-1] + dt*(eta*v*(v - v_theta)*u1)
    w2[j] = w2[j-1] + dt*(eta*v*(v - v_theta)*u2)
    
    
    w1[j] = min(w1[j], 10)
    w2[j] = min(w2[j], 10)

    pattern = pattern + dt
    if pattern > ts: 
        pattern = 0
        if flag == 1: flag = 0; 
        else: flag = 1; 

fig,axes = plt.subplots(); 

axes.plot(tvec, w1, label = "W1")
axes.plot(tvec, w2, label = "W2")
plt.legend()
plt.show()
```

When the non-silent equilibrium is unstable we obtain an outward spiral. Yes, we do get an oscillation cycle for when $u_1 > 1$.
